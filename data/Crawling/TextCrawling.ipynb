```python
# -*- coding: UTF-8 -*-
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
from bs4 import BeautifulSoup


#Colab에선 웹브라우저 창이 뜨지 않으므로 별도 설정한다.
options = webdriver.ChromeOptions()
options.add_argument('--headless')        # Head-less 설정
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
def get_대출사기형_info():

    list_url     = []     # 주소 리스트
    list_content = []     # 본문 리스트


    # 드라이버 실행
    driver = webdriver.Chrome(options = options)
    driver.implicitly_wait(3)
    sleeptime = 0.5     # 프로세스 일시 정지 시간

    for page in range(1, 20):
        url = f"https://www.fss.or.kr/fss/bbs/B0000206/list.do?menuNo=200690"
        req = requests.get(url)
        html = req.text
        soup = BeautifulSoup(html, 'html.parser')
        table = soup.select(
            '#content > div.bd-list > table > tbody > tr > td.title > a'
        )

        links = []
        for a in table:
            links.append(a.attrs['href'])

        for link in links:
            # print("https://www.fss.or.kr" + link)
            req = requests.get("https://www.fss.or.kr" + link)
            html = req.text
            soup = BeautifulSoup(html, 'html.parser')
            h2 = soup.select('#content > div.bd-view > div > div')
            for content in h2:
                # print(content.text)
                list_content.append(content.text)
    voicephishing_대출사기형 = pd.DataFrame({'내용': list_content})
    return voicephishing_대출사기형
```

```python
# 피해자/사기범 명칭 제거
for i in range(len_voicephishing_대출사기형_list):
    cur_text = voicephishing_대출사기형_list[i]
    
    for j in range(len(cur_text)):
        # 피해자 명칭 제거
        if ('피해자:' in cur_text[j]) or ('피해자 : ' in cur_text[j]):
            cur_text[j] = re.sub(r'피해자:', '', cur_text[j])
            cur_text[j] = re.sub(r'피해자 : ', '', cur_text[j])
            voicephishing_대출사기형_list[i][j] = cur_text[j]
            continue
            
        # 사기범 명칭 제거
        if ('사기범:' in cur_text[j]) or ('사기범 : ' in cur_text[j]):
            cur_text[j] = re.sub(r'사기범:', '', cur_text[j])
            cur_text[j] = re.sub(r'사기범 : ', '', cur_text[j])
            voicephishing_대출사기형_list[i][j] = cur_text[j]
        # 사기범 명칭 제거
        if ('사기범2:' in cur_text[j]) or ('사기범2 : ' in cur_text[j]):
            cur_text[j] = re.sub(r'사기범2:', '', cur_text[j])
            cur_text[j] = re.sub(r'사기범2 : ', '', cur_text[j])
            voicephishing_대출사기형_list[i][j] = cur_text[j]
        
        # 사기범 명칭 제거
        if ('사기범3:' in cur_text[j]) or ('사기범3 : ' in cur_text[j]):
            cur_text[j] = re.sub(r'사기범3:', '', cur_text[j])
            cur_text[j] = re.sub(r'사기범3 : ', '', cur_text[j])
            voicephishing_대출사기형_list[i][j] = cur_text[j]
                
```

```python
# 문장 분리
for i in range(len_voicephishing_대출사기형_list):
    cur_text = voicephishing_대출사기형_list[i]
    splited_text = []
    
    for j in range(len(cur_text)):       
        # 문장 분리 후 길이 저장
        splited = split_sentences(cur_text[j])
        len_splited = len(splited)
        
        # 길이가 1이면 바로 추가함
        if len_splited == 1:
            splited_text.append(cur_text[j])
        # 길이가 1이 아니면 나눠서 추가함
        else:
            for k in range(len_splited):
                splited_text.append(splited[k])
            
    # 분리한 문장으로 대체
    voicephishing_대출사기형_list[i] = splited_text
# 욕설 제거
for i in range(len_voicephishing_대출사기형_list):
    cur_text = voicephishing_대출사기형_list[i]
    
    for j in range(len(cur_text)):                      
        # 욕설 이후 제거
        if 'xx' in cur_text[j]:
            del voicephishing_대출사기형_list[i][j:]
            break
                   
```

```python
# 문장 분리
for i in range(len_voicephishing_수사기관사칭형_list):
    cur_text = voicephishing_수사기관사칭형_list[i]
    splited_text = []
    
    for j in range(len(cur_text)):       
        # 문장 분리 후 길이 저장
        splited = split_sentences(cur_text[j])
        len_splited = len(splited)
        
        # 길이가 1이면 바로 추가함
        if len_splited == 1:
            splited_text.append(cur_text[j])
        # 길이가 1이 아니면 나눠서 추가함
        else:
            for k in range(len_splited):
                splited_text.append(splited[k])
            
    # 분리한 문장으로 대체
    voicephishing_수사기관사칭형_list[i] = splited_text
# 욕설 제거
for i in range(len_voicephishing_수사기관사칭형_list):
    cur_text = voicephishing_수사기관사칭형_list[i]
    
    for j in range(len(cur_text)):                      
        # 욕설 이후 제거
        if 'xx' in cur_text[j]:
            del voicephishing_수사기관사칭형_list[i][j:]
            break
                   
```
